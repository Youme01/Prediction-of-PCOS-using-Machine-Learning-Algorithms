```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```
corr_features=pcos.corrwith(pcos["PCOS (Y/N)"]).abs().sort_values(ascending=False)
#features with correlation more than 0.05
corr_features=corr_features[corr_features>0.4].index
corr_features
```

```
pcos['Marraige Status (Yrs)'].fillna(pcos['Marraige Status (Yrs)'].median(),inplace=True)
pcos['Fast food (Y/N)'].fillna(pcos['Fast food (Y/N)'].median(),inplace=True)
```

```
x= pcos[['Follicle No. (R)', 'Follicle No. (L)',
       'Skin darkening (Y/N)', 'hair growth(Y/N)', 'Weight gain(Y/N)',
       'Cycle(R/I)']]
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=101,shuffle = True)
```

```
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```
from sklearn.metrics import classification_report
```

```
prediction = logmodel.predict(x_test)
print(classification_report(y_test,prediction))
```

```
from sklearn.model_selection import StratifiedKFold,KFold
```

```
skf = StratifiedKFold(n_splits=5)
skf.get_n_splits(x, y)
```

```
StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
print(skf)
```

```
for train_index, test_index in skf.split(x, y):
    print("TRAIN:", train_index, "TEST:", test_index)
X_train1, X_test1 = x.loc[train_index], x.loc[test_index]
y_train1, y_test1 = y.loc[train_index], y.loc[test_index] 

logmodel.fit(X_train1,y_train1)
  #TRAIN: [1 3] TEST: [0 2] TRAIN: [0 2] TEST: [1 3]
```

```
predictions1 = logmodel.predict(X_test1)
```

```
print(classification_report(y_test1,predictions1))
```

```
from sklearn import metrics 
from sklearn.metrics import r2_score
print('MSE_1:', metrics.mean_squared_error(y_test, prediction))
print('MSE_2:', metrics.mean_squared_error(y_test1, predictions1))
print('r2 score of holdout: ',r2_score(y_test, prediction))
print('r2 score of stratifiedkFold',r2_score(y_test1, predictions1))
```

```
from sklearn.metrics import confusion_matrix
```

```
print('HFold: ')
print(confusion_matrix(y_test,prediction))
```

```
print('SKFold: ')
print(confusion_matrix(y_test1,predictions1))
```
